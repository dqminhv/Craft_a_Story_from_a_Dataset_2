```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

# Import Data

```python
data_path = '/home/dqminhv/Springboard/Craft_a_Story_from_a_Dataset/data/Global_YouTube_Statistics.csv'
```

```python
#Import csv data file to a dataframe
try:
    # Try reading the file with UTF-8 encoding
    df = pd.read_csv(data_path, encoding='utf-8')
except UnicodeDecodeError:
    # If UTF-8 fails, try other encodings
    try:
        df = pd.read_csv(data_path, encoding='latin-1')
    except UnicodeDecodeError:
        # Handle the error or try additional encodings as needed
        print("Error: Could not read the file with any of the specified encodings.")
```

```python
#Look at the first few entries
df.head().T
```

# Handling Missing Values

```python
#Check for missing values
msno.matrix(df)
```

```python
df['category'].fillna('Other', inplace=True)
df['channel_type'].fillna('Other', inplace=True)
```

```python
df['Country'].fillna('Unspecified', inplace=True)
```

```python
df[['Gross tertiary education enrollment (%)', 'Population','Unemployment rate', 'Urban_population']] =\
df[['Gross tertiary education enrollment (%)', 'Population','Unemployment rate', 'Urban_population']].\
fillna(df[['Gross tertiary education enrollment (%)', 'Population','Unemployment rate', 'Urban_population']].mean())
```

```python
df.info()
```

# Analysis

```python
sns.barplot(data=df.groupby('category', as_index=False).agg({'uploads':'sum', 'subscribers':'sum'}),\
         x='category', y='uploads', hue='category', legend=False)
plt.xticks(rotation=90)
```

## Top categories

```python
sns.countplot(x='category', data=df, stat='percent', order=df['category'].value_counts().index,\
              hue='category', legend=False)
plt.xticks(rotation=90)
plt.show()
```

## Number of subscriber for each chanel category

```python
sns.barplot(data=df, x='category', y='subscribers', hue='category', legend=False)
plt.xlabel('Chanel Category')
plt.xticks(rotation=90)
plt.ylabel('Number of subscriber')
```

## Top countries

```python
sns.barplot(data=df, x='Country', y='subscribers', hue='Country', legend=False)
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Number of subscriber')
```

### By the number of channels

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Country', data=df, order=df['Country'].value_counts().index,\
             stat='percent', hue='Country', legend=False)
plt.xticks(rotation=90)
plt.show()
```

### Account for the population

```python
#Create a dataframe that count the number of Youtube channels for each country
country_count = pd.DataFrame(df['Country'].value_counts())
```

```python
#Create a population dataframe
population = df[['Country', 'Population']].groupby('Country').agg({'Population':'first'})
```

```python
#Merge two dataframes
result_df = pd.merge(country_count, population, left_index=True, right_index=True).reset_index()
```

```python
#Calculate the number of youtube channels per 1 millions people
result_df['PerMi'] = result_df['count'] / result_df['Population'] * 1000000
```

```python
result_df.head()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Country', y='PerMi', data=result_df.sort_values('PerMi', ascending=False), hue='Country', legend=False)
plt.xticks(rotation=90)
plt.show()
```

### Account for the urban population

```python
urban_population = df[['Country', 'Urban_population']].groupby('Country').agg({'Urban_population':'first'}).reset_index()
```

```python
urban_population.head()
```

```python
#Merge two dataframes
result_df = pd.merge(result_df, urban_population, on='Country').reset_index()
```

```python
#Calculate the number of youtube channels per 1 millions people
result_df['Per_Mi_Urban'] = result_df['count'] / result_df['Urban_population'] * 1000000
```

```python
result_df.head()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Country', y='Per_Mi_Urban', data=result_df.sort_values('Per_Mi_Urban', ascending=False), hue='Country', legend=False)
plt.xticks(rotation=90)
plt.show()
```

## Categories in each country

```python
country_category = df[['Country','category']]
```

```python
country_category.head()
```

```python
for country in country_category['Country'].unique():
    plt.figure(figsize=(8,6))
    sns.countplot(data=country_category[country_category['Country']==country], x='category',\
                  hue='category', legend=False,\
                  order=country_category[country_category['Country']==country]['category'].value_counts().index)
    plt.xlabel('Category')
    plt.xticks(rotation=90)
    plt.title('{}'.format(country))
    plt.show()
    plt.close()
```
